```python
df['SMA_12'] = df['EnergyIndex'].rolling(window=12).mean()
```

```python
from pylab import rcParams

rcParams['figure.figsize'] = (12,5)
```

```python
df.plot()
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

span= 12
alpha = 2 / (span+1)

model1 = SimpleExpSmoothing(df['EnergyIndex'])

fittedmod1 = model1.fit(smoothing_level = alpha, optimized = False)

df['SES_12'] = fittedmod1.fittedvalues.shift(-1)
```

```python
df.plot();
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
model2 = ExponentialSmoothing(df['EnergyIndex'], trend='mul', seasonal='mul', seasonal_periods=12)
```

```python
fitted_model2 = model2.fit()
```

```python
df['TES_mul_12'] = fitted_model2.fittedvalues.shift(-1)
```
