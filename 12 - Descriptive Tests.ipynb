```python
# interpreting these results of ADF test
```

p-value is imp as its the indicator of whether the test fails or holds.

<b> its better to make a func of adf test that takes in a series and gives you directly if its stationary or not, rather than reading p-values each time.</b>
    
    > data with a UNIT ROOT is NOT STATIONARY

```python
# defining a func for ADF results

from statsmodels.tsa.stattools import adfuller

def adf_test(series, title=''):

    """
    Pass in a TS and an optional Title, returns an ADF report
    """
    
    print(f'ADF Test for : {title}')
    
    result = adfuller(series.dropna(), autolag='AIC')   
    #dropna() drops NaN containg rows from differencing step
    
    labels = ['ADF test stat', 'p-value', 'lags used', 'observations']
    out = pd.Series( result[0:4], index=labels )
    
    for key, val in result[4].items():
        out[f'critical value ({key})'] = val
        
    print(out.to_string())   # to_string() removes line 'dtype : float64'
    
    
    # H0 = Non-Stationary dataset
    
    if result[1] <= 0.05 :#p-value
        print('reject H0 Null Hypothesis\n')
        print('Data has NO UNIT ROOT\n')
        print('STATIONARY')
        
    else :
        print('\n')
        print('accept H0 Null Hypothesis')
        print('Data has a UNIT ROOT\n')
        print('Non - Stationary')
    
```

> so all we care abt is the P-VALUE , second returned value [1]

```python
# passing our df col into the function above and see what we get

adf_test( df1 ['Thousands of Passengers'], title='airline passengers' )
```

```python
# in the plot its evident that the data is non-stationary

df1['Thousands of Passengers'].plot(color='red', figsize=(8,5));
```

# Granger Causality test

causality : when one var changes wrt other and can be used to forecast the other one.

causality doesn't imply correlation.

even if GCT returns the indication of causality, it might be untrue.

```python
df3 = pd.read_csv('Data/samples.csv', parse_dates=True, index_col=0)
```

```python
df3.head()
```

```python
# to use statsmodels

df3.index.freq = 'MS'
```

```python
from matplotlib.pyplot import rcParams

rcParams["figure.figsize"] = (12,6)
```

```python
# plotting 'a' vs 'd'

df3[['a','d']].plot();
```

```python
#lets zoom in to see if there's causality

df3['a'].iloc[2:].plot(legend=True, title='column-a');

df3.d.shift(2).plot(legend=True);
```

we're getting almost clear alignment here. there's some noise but still v close.

as they're lining on top each other so we can expect a relationship here.

<b> we shifted 'd' by 2 and then plotted it with original 'a'</b>

so what does the above graph tells ?

> that 2 Days after a change in 'd', that change reflects in 'a' => there might be some causality b/w these two

these couldn't be seen w naked eye, and we've clearly been v lucky in guessing the shift by 2 units. but you might not be v lucky.

> USE GCT INSTEAD ! TO CHECK FOR CAUSALITY

```python
from statsmodels.tsa.stattools import grangercausalitytests

help(grangercausalitytests)
```

```python
grangercausalitytests(df3[['a','d']], maxlag=3)
```

parameters :

1. maxlag : shifting amount . <br>
    difficult to guess by yourself.
    
    more maxlags => greater computational time <br>
    less maxlags => lower possibility of detecting causality
    
    
output :

> look for very low p-values 

```python
# lets compare for another set of columns that have no causality 

# we know b & d cols have no causality

df3[ ['b', 'd'] ].plot();
```

```python
# no causality evident w eyes

#lets use GCT on b & d to confirm
```

```python
grangercausalitytests(df3[['b','d']], maxlag=3)
```

<br>
> <b> p-val = 0.05 : cut-off for any hypotheses test
    
</b>

so :

* p-val < 0.05   => Causality present
* p-val > 0.05   => Causality absent

in results for 'b & d', all p-val are greater than 0.05 so there's no Causality

so in GCT, pass 2 cols of df[['c1', 'c2']], check for p-val. lower than 0.05 mean Causality is there.

> lag values at which causality is strong have significance more than just causality indication, so dive further into data.

eg: for monthly sales data, if lag=12 has strong causality means a year later it has a strong cause on the other dataset col.

<b> so GCT gives you lag values to explore further and get more info abt data. like hey this lag has strong causality so go ahead and understand its meaning.

# using EM for Forecasts

```python
# creating an imaginary dataset

np.random.seed(42)

df = pd.DataFrame( np.random.randint(20,30, (50,2)), columns=['test', 'preds'] )
```

```python
df.head()
```

```python
df.plot();
```

```python
#from sklearn.metrics import mean_squared_error

#statsmodels also has a lot of Evaluation Metric including MSE, BIC, AIC

from statsmodels.tools.eval_measures import mse, rmse, meanabs
```

```python
# the order of passing doesn't matter coz you take abs value of diff or square it
# so order doesn't matter

mse(df['test'], df['preds'], axis=0)
```

```python
rmse(df['test'], df['preds'])
```

```python
meanabs(df['test'], df['preds'])
```

generally rmse is prepared as the EM coz:

1. it punishes the model a lot for larger errors, as its root of mse.
2. has same units as original data

```python
from statsmodels.graphics.tsaplots import month_plot, quarter_plot
```

what these do?

when you've timestamped index, these plots can separate values to months & quarters.

```python
month_plot( df['Thousands of Passengers']);
```

```python
# all values that fall in JAN, irrespective of the year

df[df.index.month==1]
```

what month_plot() does?

1. groups by month name. ie jan datapoints are grouped together, feb ones together.
<br>
2. sets xticks = month names
<br>
3. for each xtick = month, plots all values for Jan for each year.

<br>
red marks = avg for that month over all years


IN THE ABOVE MONTH_PLOT :

july, aug have higher red ticks ie greater avg values for summer month which is the actual case for this dataset.
