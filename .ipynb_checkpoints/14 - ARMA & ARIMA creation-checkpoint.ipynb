###### CLASSICAL APPROACH FOR p,d,q

```python
from statsmodels.tsa.stattools import adfuller

from statsmodels.tsa.statespace.tools import diff
```

```python
adfr = adfuller(df2['Inventories'])
```

```python
adfr
```

```python
# H0 : unit root exists => non-stationary

if adfr[1]< 0.05 :
    print('Stationary')
    
else:
    print('Non-Stationary')
```

```python
df2['DIFF1'] = diff(df2['Inventories'], k_diff=1)
```

```python
adfr2 = adfuller(df2['DIFF1'].dropna(axis=0))  #axis=0 : rows
```

```python
adfr2
```

```python
# H0 : unit root exists => non-stationary

if adfr2[1]< 0.05 :
    print('Stationary')
    
else:
    print('Non-Stationary')
```

```python
# one difference made it stationary

# plotting ACF & PACF plots :
```

```python
import statsmodels.api as sm

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# ACF for MA-model order (q)

plot_acf(df2['Inventories'], lags=40);

#gradual decrease w each being equally significant
```

```python
# PACF for AR-order

plot_pacf(df2['Inventories'], lags=40);
```

Through CLASSICAL METHODS to choose p,d,q :

1. d : by diff() till you get stationary data, then ADF-test

2. p : above PACF plot has a steep cut off => use AR component w order = lag at which blue area cuts. here p=1

3. q : MA = 0 OR 1 #check which one has less error

SPLITTING DATASET :

```python
len(df2)
```

```python
# last year for test - 12 units

test = df2.iloc[252:]

train = df2.iloc[:252]
```

```python
model = ARIMA(train['Inventories'], order=(1,1,1))
```

```python
fitted_model = model.fit()
```

```python
fitted_model.summary()
```

param 'typ'= :

remember, here we're using a differenced component containing model - ARIMA .

1. 'linear' : retuns PREDICTIONS differenced by 1 

2. 'levels' : returns prdictions off of originak vars

<b> do we want the values to come off from the original TS or the differenced TS ?
    
</b>

we dodn't have to worry abt this, coz till now we used model w/o the differencing component.


> for model with a differencing step (I-component)  :  use typ='levels' in .predict()

```python
star =len(train)
en = len(train) + len(test) - 1

preds = fitted_model.predict(start=star, end=en , typ='levels').rename('ARIMA 1,1,1 PREDS')
```

```python
preds.head()
```

```python
test.head()
```

if typ wouldn't be 'levels', the fitting would be done wrt 'DIFF1' column above, not Inventories like we want.

```python
# plotting test actual vs preds

test['Inventories'].plot(legend=True);

preds.plot(legend=True);
```

we had a small seasonal component that we decided is not useful, and hence used ARIMA instead of SARIMAX.

> so our preds are showing just the GENERAL TREND, not the seasonality as it was not taken into account while building this model. 

but is it really that our model ain't picking the seasonality? can't say this Zoomed-in to the output.

> comparing AVERAGES and RMSE to evaluate the model against the test set

```python
test['Inventories'].mean()  #1
```

```python
preds.mean()

# both the means are pretty close yk
```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
error = rmse(test['Inventories'], preds)
```

```python
error  #2
```

```python
test['Inventories'].std()
```

comapare rmse with avg value ie #1 nd #2.

you see error is not v significant wrt average value for test set.

order of thousands vs order of millions

> so we're overpredicting & underpredicting at times, but the obtained general trend is good.

###### forecasting the actual future
