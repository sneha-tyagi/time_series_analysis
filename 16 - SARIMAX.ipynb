### Running SARIMA only at first

TRAIN TEST SPLIT AND EVALUATION

```python
len(df)
```

```python
len(df) - 42  # 42 days
```

```python
train = df.iloc[:436]

test = df.iloc[436 :]
```

```python
# find orders of SARIMA

from pmdarima import auto_arima

auto_arima(df['total'], seasonal=True, m=7).summary()
```

```python
# creating & fitting model on these orders

from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(train['total'] , order=(1,0,0), seasonal_order=(2,0,0,7), enforce_invertibility=False)
```

parameter : <b>enforce_invertibility = False </b>

* this additional is specified due to the way statsmodels internally processes SARIMA model.

* invertibility : for an invertible process, most recent obsv have greater weight than the past ones, which is preferred.

* but SARIMA internally can get you an error

<b>value error : non-invertible starting MA paramters found </b>

> if you get an error, try 'enforce_invertibility = False' and the error should disappear.

```python
fm = model.fit()
```

```python
fm.summary()
```

```python
# predicting test set

start = len(train)

end = len(train) + len(test) - 1

preds1 = fm.predict(start, end, typ='levels').rename('SARIMA')
```

```python
# plotting test actual vs predicted

test['total'].plot(legend=True);

preds1.plot(legend=True);
```

we're doing just okay...

lets see at places where we're very off,if those days had holidays. plot our holidays and see if they're are affecting our predictions.

```python
_ = test['total'].plot(legend=True);

preds1.plot(legend=True);

for day in test.query('holiday==1').index:
    _.axvline(x=day, color='black')
```

```python
# evaluating above model

from statsmodels.tools.eval_measures import rmse
```

```python
# we'll compare this error with that of SARIMAX to see the quatitative differences in their performaces.
```

```python
error1 = rmse(test['total'], preds1)

error1
```

```python
test['total'].mean()
```

```python
# large error
```

# SARIMAX to improve above model

adding an X var to a SARIMA model.

ie 'holiday' column here, as it might be affecting our preds. so lets add it and see if model improves.

```python
# passing in X var holiday 

r = auto_arima(df['total'], exogenous = df[['holiday']], seasonal=True, m=7)
```

exogenous param : accepts DATAFRAME , not a single Series / column. 

(thats the way its internally built in statsmodels)

> you can pass multiple columns in 'exogenous' as a dataframe

<br>


but here we've 1 column/series, what to do ?
* <b> just add an extra set of [] to it so that it becomes a dataframe.
    
    dataframe[ ['col'] ] </b>


```python
r.summary()
```

see after adding X var we're getting diff order in auto_arima()

#### train our sarimax

on train test split procedure followed earlier.

```python
model = SARIMAX( train['total'], exog = train[['holiday']], order=(1,0,1), seasonal_order=(1,0,1,7), enforce_invertibility=False)
```

```python
fm = model.fit()
```

```python
# predicting

start = len(train)
end = len(train) + len(test) - 1
```

see we're passing in X var of test set for predictions , why?

> model takes in its algo + the values of X for a future-ts > returns value of TV for that future-ts

ie we gotta<b> pass the X var for every future date to get TV.

```python
preds = fm.predict(start, end, exog=test[['holiday']], typ='levels').rename('SARIMAX')
```

```python
# plotting preds vs test actual

_ = test['total'].plot(legend=True);

preds.plot(legend=True);

#preds1.plot(legend=True);

# plotting HOLIDAYS too using query() substitute
for x in test[test['holiday']==1].index:
    _.axvline(x=x, color='k', alpha=0.8)
```

we seem to be doing pretty well for holidays, but we might be a little off for non-holidays, than before.

```python
# evaluation metric - quatifying measure for the error

errorx = rmse(test['total'], preds)
```

```python
errorx
```

```python
# significantly less than 41.2 of SARIMA
```

ques to ask ourselves :

> does RMSE value of SARIMAX has a sugnificant improvement than SARIMA so that its worth it to add that 'X' ? 

ie we'll compare the two models using a evaluation metric - rmse !

here, we're halving the error so its great to add X.
